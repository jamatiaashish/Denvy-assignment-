```python
import pandas as pd

# Load dataset
df = pd.read_excel("SET-8.xlsx")  # Make sure filename matches
print("Dataset:")
print(df)
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Drop if any
df = df.dropna()
```

```python
# Define independent features (X)
X = df[['ground_clearance', 'height']]
print("Independent Features (X):")
print(X.head())  # show first 5 rows for clarity

# Define dependent feature (target variable)
y = df['category']
print("\nTarget (y):")
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 70-30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Train model
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)
```

```python
# Predict and check accuracy
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy (70-30 split): {accuracy:.2f}")
```
